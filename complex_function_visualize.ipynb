```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
```

## 複素関数を可視化することを目標にする。
### 教科書にはあっさり図だけが書かれていることが多いので、ここではいくつかの$f(z)$を見てみよう。

### (具体的には、$f(z)=f(x+iv)=u+iv$,$z=x+iy, w=u+iv$の曲面を観察することでf(z)の等高線を可視化する。)

### (e.g.)  $f(z)=z^2$
#### $z^2=(x+iy)^2=x^2-y^2+2ixy=u+iv$より、$u=x^2-y^2, v=2xy$となる。以下この$u,v$をプロットすると、

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = y = np.arange(-15, 15, 0.5)
X, Y = np.meshgrid(x, y)

Z = X**2 -Y**2
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.cool)
ax.tick_params(labelbottom=False,labelleft=False,labelright=False,labeltop=False)
ax.set_title('$u=x^2-y^2$', loc='right')
#plt.savefig("test.png")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = y = np.arange(-15, 15, 0.5)
X, Y = np.meshgrid(x, y)

Z = 2*X*Y
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.spring)
ax.tick_params(labelbottom=False,labelleft=False,labelright=False,labeltop=False)
ax.set_title('$v=2xy$', loc='right')
#plt.savefig("test.png")
plt.show()
```

#### これの等高線をplotすると、よく教科書で見る$f(z)$の等高線を得ることができる。

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = y = np.arange(-15, 15, 0.5)
X, Y = np.meshgrid(x, y)

Z1 = X**2 -Y**2
Z2 = 2*X*Y
ax.contour(X, Y, Z1, cmap=cm.cool, levels=10)
ax.contour(X, Y, Z2, cmap=cm.spring, levels=10)
ax.tick_params(labelbottom=False,labelleft=False,labelright=False,labeltop=False)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = y = np.arange(-15, 15, 0.5)
X, Y = np.meshgrid(x, y)

Z1 = X**2 -Y**2
Z2 = 2*X*Y
ax.contour(X, Y, Z1, colors = "red", offset = -1)
ax.contour(X, Y, Z2, colors = "blue", offset = -1)
ax.tick_params(labelbottom=False,labelleft=False,labelright=False,labeltop=False)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.set_aspect('equal', adjustable='box')
x = np.arange(-25, 25, 0.1) # x軸
y = np.arange(-25, 25, 0.1) # y軸
 
X, Y = np.meshgrid(x, y)
Z1 = X**2 - Y**2
Z2 = 2*X*Y

cont = plt.contour(X,Y,Z1,colors=['red'])
cont = plt.contour(X,Y,Z2,colors=['blue'])
ax.set_title('$f(z)=z^2$', loc='right')
ax.tick_params(labelbottom=False,labelleft=False,labelright=False,labeltop=False)
plt.show()
```

### (e.g.)  $f(z)=1/z$
#### $z^{-1}=(x+iy)^{-1}=u+iv$より、$u=x/(x^2+y^2), v=-y/(x^2+y^2)$となる。以下この$u,v$をプロットすると、

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = y = np.arange(-3, 3, 0.21)
X, Y = np.meshgrid(x, y)

Z =  X/(X**2 + Y**2)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.cool)
ax.tick_params(labelbottom=False,labelleft=False,labelright=False,labeltop=False)
ax.set_title('$u=x/(x^2+y^2)$', loc='right')
#plt.savefig("test.png")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = y = np.arange(-3, 3, 0.21)
X, Y = np.meshgrid(x, y)

Z =  -Y/(X**2 + Y**2)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.spring)
ax.tick_params(labelbottom=False,labelleft=False,labelright=False,labeltop=False)
ax.set_title('$v=-y/(x^2+y^2)$', loc='right')
#plt.savefig("test.png")
plt.show()
```

#### 同様にこれの等高線をplotすると、よく教科書で見る$f(z)$の等高線を得ることができる。

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = y = np.arange(-3, 3, 0.21)
X, Y = np.meshgrid(x, y)

Z1 =   X/(X**2 + Y**2)
ax.contour(X, Y, Z1, cmap=cm.cool, levels=50)
ax.set_title('$u=x/(x^2+y^2)$', loc='right')
ax.tick_params(labelbottom=False,labelleft=False,labelright=False,labeltop=False)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = y = np.arange(-3, 3, 0.21)
X, Y = np.meshgrid(x, y)

Z2 =   -Y/(X**2 + Y**2)
ax.contour(X, Y, Z2, cmap=cm.spring, levels=50)
ax.set_title('$v=-y/(x^2+y^2)$', loc='right')
ax.tick_params(labelbottom=False,labelleft=False,labelright=False,labeltop=False)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = y = np.arange(-100, 100,  0.22)
X, Y = np.meshgrid(x, y)

ax.set_xlim(-1.0, 1.0)
ax.set_ylim(-1.0, 1.0)

Z1 =   X/(X**2 + Y**2)
Z2 =   -Y/(X**2 + Y**2)
ax.contour(X, Y, Z1, colors = "red", offset = -1)
ax.contour(X, Y, Z2, colors = "blue", offset = -1)
ax.tick_params(labelbottom=False,labelleft=False,labelright=False,labeltop=False)
plt.show()
```
